# Rise in the CO2 Levels
This notebook will do some analysis on the co2_monthly dataset and will try to predict **whether our world will face severe cosequences with the rising level of co2**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('co2_monthly.csv')
```

```python
df.head()
```

```python
df.drop(['datenum','month','day','datetime'],axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.distplot(df['data_mean_global'],bins=50,)
```

```python
df1 = df.groupby('year')
```

```python
df1
```

```python
df1 = df1.mean()
```

```python
df1.head()
```

```python
y = df1.index
```

```python
df1['year'] = range(2015)
```

```python
df1.head()
```

```python
sns.lmplot(x='year',y='data_mean_global',data=df1,size=15,fit_reg=False,)
plt.ylabel('ppm of CO2')
```
